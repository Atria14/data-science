```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generujemy sobie dane
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)

plt.plot(x, x_sin)
```

```python
plt.style.use('ggplot') # styl gg plot
plt.plot(x, x_sin)
```

```python
# lista dotÄ™pnych styli plt.style.available
plt.style.available
```

```python
plt.style.use('seaborn-ticks')
plt.plot(x, x_sin)
```

```python
plt.style.use('dark_background')
plt.plot(x, x_sin)
```
