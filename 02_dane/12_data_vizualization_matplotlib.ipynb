```python
import matplotlib
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# podstawowym obiektem jest figure
fig = plt.figure()
fig.show()
```

Wykres liniowy

```python
plt.plot([0,1], [0,3]) # tworzy wykres
```

```python
_ = plt.plot([0,1], [0,3]) # przypisanie zmiennej powoduje usunięcie napisu nad wykresem
```

```python
plt.plot([0,1], [0,3], label='line', color='green') # zmiana akolorów
# dodanie osi X, Y
# podstawowy jest figure do którego doadjemy kolejne warstwy
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres') # nazwa wykresu
_ = plt.legend() # wyświetli legendę
```

```python
plt.plot([0,1,2], [0,3,2], label='line', color='green') # zmiana akolorów
# dodanie osi X, Y
# podstawowy jest figure do którego doadjemy kolejne warstwy
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres') # nazwa wykresu
_ = plt.legend() # wyświetli legendę
```

```python
x = np.arange(-3, 3, 0.1) # tworzy listę punktów o równym rozstawie od -3 do 3 o skoku 0.1
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin) # na osi X bedą punkty które wygenerowaliśmy a na osi Y sinus
_ = plt.plot(x, x_cos)

#po wprowadzeniu drugiego wykresu automatycznie dwa wykresy zostają na siebie nałożone
```

Tworzenie wielu wykresów

```python
plt.subplot(121) # 1 - podaje liczbę wierszy ile chcemy uzyskać w siatce wykresu, 2 - liczba kolumn, 1 - odnosi się do konkretnego wykresu, pierwszy po lewej stronie
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122) # ostatnia cyfra 2 numeruje nam wykres
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
```

```python
fig = plt.figure(figsize=(12, 8)) # polecenie określa wielkość wykresy w pikselach

# utowrzymy 3 wykesy w wierszu

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132) 
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133) 
plt.plot([0, 1, 2], [3, 1, 4], 'ro')

fig.show() # w pycharm trzeba dodać tą metodę by wykres został wyświetlony
```

Wykresy słupkowe

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])

# tak samo jak w liniowych można dodawać właściwości, zmieniać kolory itd
```

```python
# definicja wartości osi x

plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
# tworzenie wykresu słupkowego horyzontalnego
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
```

Wykresy punkotwe

```python
x1 = np.random.random(300)
x2 = np.random.random(300)
                      
# wykres scatter
plt.scatter(x=x1, y=x2)
```

```python
# zmiana rozmiaru kropki
x3 = np.random.random(300)
plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5) # parametr alpha oznacza przezroczystość kropek
```
