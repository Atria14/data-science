```python
import numpy as np
import pandas as pd
```

WCZYTANIE DANYCH I PREPROCESSING

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

1. ZOSTAWIAMY SUROWE DANE ABY MOŻNA BYŁO DO NICH WRÓCIĆ, TWORZYMY KOPIĘ:

```python
retail = retail_raw.copy()
retail.info() # wyświetlamy info
# CustomerID - nie chcemy aby było float
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail.Quantity >= 0 # warunek logiczny pokazujący nam tylko dodatnie Quantity
# wycinamy warunek, usunięte zostaną wartości z ujemnym Quantity
retail = retail[retail.Quantity >= 0] # nadpisujemy zmienną retail
retail.info()
```

```python
# zmieniamy CustomerID z float64 na object
# krok 1 - przekonwertować float64 na liczbę całkowitą int a następnie na str
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```

PORÓWNANIE DO JĘZYKA SQL

```python
# SELECT * FROM retail;
retail
```

```python
# SELECT Quantity, UnitPrice, Customer ID FROM retail;
retail[['Quantity', 'UnitPrice', 'CustomerID']]
```

```python
# SELECT Quantity, UnitPrice, Customer ID FROM retail LIMIT 10;
retail[['Quantity', 'UnitPrice', 'CustomerID']][:10]
```

```python
# SELECT * FROM retail WHERE CustomerID='17850';
retail[retail.CustomerID == '17850']
```

```python
retail.query('CustomerID == "17850"') #druga metoda, trzeba uważać na cudzysłowia
```

```python
# SELECT * FROM retail WHERE CustomerID='17850' and UnitPrice > 5;
retail[(retail.CustomerID == '17850') & (retail.UnitPrice >=5)]
```

```python
# SELECT * FROM retail WHERE CustomerID='17850' or Country='France';
retail[(retail.CustomerID == '17850') | (retail.Country == 'France')]
```

```python
# SELECT * FROM retail WHERE InvoiceNO is not null;
retail.InvoiceNo.notnull()
retail[retail.InvoiceNo.notnull()]
```

```python
# SELECT CustomerID, count(*) FROM retail GROUP BY CustomerID;
retail.groupby('CustomerID').size()
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM retail GROUP BY CustomerID;
# najpierw musimy utworzyć kolumnę Revenue
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size})
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}). \
  rename(columns={'Revenue': 'RevenueAvarage', 'CustomerID': 'CustomerIDCount'})
```

```python
# SELECT InvoiceDateDay, sum(Revenue) FROM retail GROUP BY InvoiceDateDay;
```

```python
retail.head(2)
```

```python
# tworzymy nową kolumnę, w której będzie dzień
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day # dt - funkcja date time, po jej zastosowaniu możemy wybrać dzień
retail.head(2)
```

```python
results = retail.groupby('InvoiceDateDay').aggregate({'Revenue': np.sum}) # za pomocą funkcji aggregate sumujemy przychód np.sum
results
```

```python
results.plot(kind='bar')
```
