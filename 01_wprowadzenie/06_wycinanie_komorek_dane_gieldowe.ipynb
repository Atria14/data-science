```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq') # biblioteka DataReader - pozwala ściągnąć dane

df = fetch_financial_data()
df.info()
```

```python
# wycinanie metodą iloc, która przyjmuje indeksy : - wszystkie indeksy, jeśli chcemy dwie kolumny to tworzymy wewnątrz listę [0,2] i podajemy ich indeksy
df.iloc[:, [0,3]]
```

```python
df.iloc[:, -1] # Series
```
